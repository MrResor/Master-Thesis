```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('bays29.points') as f:
    lines = f.readlines()
lines = lines[1:]
```

```python
nodes = len(lines)
points = np.zeros((nodes, 2))
for i, line in enumerate(lines):
    points[i] = list(map(float, line.strip().split(",")[1:]))    
```

```python
for ind, p1 in enumerate(points):
    for p2 in points[ind+1:]:
        plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color='blue', zorder=1)
plt.scatter(points[:, 0], points[:, 1], color='red', zorder=2)
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])
plt.show()
```
