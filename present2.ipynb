```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('bays29.points') as f:
    lines = f.readlines()
lines = lines[1:]
```

```python
nodes = len(lines)
points = np.zeros((nodes, 2))
for i, line in enumerate(lines):
    points[i] = list(map(float, line.strip().split(",")[1:]))    
```

```python
for ind, p1 in enumerate(points):
    for p2 in points[ind+1:]:
        plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color='blue', zorder=1)
plt.scatter(points[:, 0], points[:, 1], color='yellow', edgecolors='black', zorder=2)
plt.axis('off')
plt.show()
```

```python
plt.plot([0, 0], [0, 1], color='blue', zorder=1)
plt.plot([0, 1], [0, 0], color='blue', zorder=1)
plt.plot([0, 1], [1, 1], color='blue', zorder=1)
plt.plot([1, 1], [1, 0], color='blue', zorder=1)
plt.scatter([0, 0, 1, 1,], [0, 1, 0, 1], color='yellow', edgecolors='black', zorder=2)
plt.text(0.01, 0.02, "0")
plt.text(0.01, 1.02, "1")
plt.text(1.01, 0.02, "2")
plt.text(1.01, 1.02, "3")
plt.axis('off')
plt.show()
```
