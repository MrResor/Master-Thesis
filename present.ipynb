```python
data = pd.read_csv('xqg237.csv').values
present = np.zeros((2 * data.shape[0], data.shape[1]))
for i, d in enumerate(data):
    present[2*i] = d
    present[2*i + 1] = [d[1], d[0], d[2]]
present = pd.DataFrame(present)
present.columns = ['From', 'To', 'Distance']
print(present.to_string(index=False))
```
